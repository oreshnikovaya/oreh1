Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%%sql
select t.name
from tracks as t, employees as e, customers as c, invoices as i, invoice_items as inv
where e.EmployeeId = SupportRepId and c.CustomerId = i.CustomerId and i.InvoiceId = inv.InvoiceId
and inv.TrackId = t.TrackId and e.HireDate >= '2002-09-01'
```

## Задание 2
Вывести общую стоимость и продолжительность каждого плейлиста.

```python
%%sql
select p.Name, round(sum(t.UnitPrice),2) as Cost, sum(t.Milliseconds)/1000 as 'Time(sec)'
from playlists as p, playlist_track as pt, tracks as t
where t.TrackId = pt.TrackId and pt.PlaylistId = p.PlaylistId
group by p.Name
```

## Задание 3
Для каждого артиста определить все жанры песен, которые у него есть и количество песен по каждому жанру.


```python
%%sql
select ar.Name, g.Name, count()
from artists as ar, tracks as tr, albums as al, genres as g
where g.GenreId = tr.GenreId and tr.AlbumId = al.AlbumId and al.ArtistId = ar.ArtistId
group by ar.Name, g.Name
```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру рок

```python
%%sql
select c.Company, count()
from genres as g, tracks as t, invoices as inv, invoice_items as invit, customers as c
where g.GenreId = t.GenreId and t.TrackId = invit.TrackId and invit.InvoiceId = inv.InvoiceId and
inv.CustomerId = c.CustomerId and g.Name = 'Rock'
group by c.Company
```
